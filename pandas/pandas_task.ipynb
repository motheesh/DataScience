```python
import pandas as pd

data_df=pd.read_table("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv")


#create new column for price without dollar symbol 
data_df["price"]=data_df["item_price"].replace('[\$,]', '', regex=True).astype(float)
#data_df["price"]=data_df["price"].astype(str).astype(int)
print(data_df["price"].dtypes)

#find sum of price
print("sum of Price: ",sum(data_df["price"]))

#find max price
print("max price : ",max(data_df["price"]))
#get the min price
print("min Price: ",min(data_df["price"]))


#get item_name with heighest price
print("max price : ",data_df[["item_name","price"]][data_df["price"]==max(data_df["price"])])


#get item_name with lowest pric
min_price=data_df[["item_name","price"]][data_df["price"]==min(data_df["price"])]
print("item with min price : ",min_price.iloc[0].price, " ",min_price.iloc[0].item_name)


#how many unique items present
print("how many unique items present",len(data_df["item_name"].unique()))

#get the total product worth of all items

print("Total Product Worth is :",(sum(data_df["quantity"]*data_df["price"])))

#how many items we have below given range of price 
price_range=30
print(f"items we have below {price_range} range of price is :",len(data_df["item_name"][data_df["price"]<price_range].unique()))


#how many items we have above the given range of price 
price_range=30
print(f"items we have above {price_range} range of price is :",len(data_df["item_name"][data_df["price"]>price_range].unique()))

#get the average price of all items
print("average price of all items",data_df.describe()["price"]["mean"])


#get all details of item which is having maximum price

print("get all details of item which is having highest maximum price")
data_df[data_df["price"]==max(data_df["price"])]
#data_df.describe()
```
